```python
plt.figure(figsize=(10,4), dpi=80)
menu_category = mcdonald.Category.value_counts()
menu_category.plot.bar(color = ['green','red','red','red','red','red','red','red','red'])
plt.title("Number of Menu Items for each Food Category")
plt.ylabel("Count")
plt.xlabel("Menu Category")
plt.xticks(rotation=45)
plt.show()
```

```python
mcdonald=mcdonald.groupby(['Category']).mean()['Sugars']
mcdonald.plot(kind='bar', figsize=(10,5))
```

As we can see from findings that smoothiness and shakes hava the highest amount of sugar. The second highest category comes under coffee and tea.Below this beverages and desserts has similar numbers and more than the remaining categories like salads and breakfast.The finsings simply support the fact that beverages generally contain the more sugar.

```python
mcdonald=mcdonald.groupby(['Category']).mean()['Protein']
mcdonald.plot(kind='bar',figsize=(15,5),grid=True)
```
